```python
#Scale data and add noise to a target for simple regression
def get_syntetic_target(x):
     noise = np.random.rand(len(x)) *10
     return (x * 100) + noise

#Generate a random dataset
def get_random_dataset(total_train, total_test):
     x_train = np.random.rand(total_train)
     y_train = get_syntetic_target(x_train)
     
     x_test = np.random.rand(total_test)
     y_test = get_syntetic_target(x_test)
     
     return x_train, y_train, x_test, y_test 
```

Generate and plot dataset for example

```python
x_train, y_train, x_test, y_test = get_random_dataset(300, 30)
```

```python
#Plot distribution    
def plot_distribution(x, y, title):
    plt.figure()
    #scatter
    plt.title(title)
    plt.scatter(x, y)
    plt.show()

#Show distributiion for training data
plot_distribution(x_train, y_train, "Training data")

#Show distribution for test data
plot_distribution(x_test, y_test, "Test data")
```

```python
#Loss function for linear regression model
def mean_squared_loss(y, y_prime):
     return np.mean((y_prime - y)**2)
```

```python
#Function to measure accuracy of the model 
def R2Coefficient(y, predictions):
     #Squared residuals
     res =  np.sum((y - predictions)**2) 
     tot = np.sum((np.mean(y) - y) **2 )

     return 1 - (res / tot)
```

```python
#Build multiple linear regression model
model = LinearRegression(loss_function=mean_squared_loss, accuracy_function=R2Coefficient)
```

```python
#Train model
model.train(x_train, y_train, True, epochs=300)
```

```python
#plot line per feature
def plot_loss_accuracy(history, epochs):
    plt.figure()
    fig, ax = plt.subplots(1, 2, figsize=(25,25))
    fig.tight_layout()
    
    x_line = np.arange(0, epochs, 1)
    
    ax[0].set_title("Train/Test Losses")
    ax[0].plot(x_line, history['training']['loss'], color='blue')
    ax[0].plot(x_line, history['validation']['loss'], color='green')
    ax[0].legend(["Training loss", "Validation loss"])

    ax[1].set_title("Train/Test Accuracy")
    ax[1].plot(x_line, history['training']['accuracy'], color='orange')
    ax[1].plot(x_line, history['validation']['accuracy'], color='red')
    ax[1].legend(["Training accuracy", "Validation accuracy"])
```

# Logistic Unit

```python
from logistic_unit import LogisticUnit
from sklearn.preprocessing import LabelBinarizer
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-1*x))
```

```python
#Plot for sigmoid function 
indexes = np.arange(-6, 6, 0.1)

plt.figure()
plt.plot(indexes, sigmoid(indexes))
plt.title("Sigmoid plot")
plt.show()
```

```python
def binary_cross_entropy(Y, predicted):
    return (-Y * np.log(predicted) - (1 - Y) * np.log(1 - predicted))
```

```python
#Test binary cross-entropy
np.mean(binary_cross_entropy(np.array([1, 0, 1]), np.array([0.1, 0.8, 0.9])))
```

```python
def get_class(value):
    if value > 0.5: 
        return 1
    
    return 0

def binary_acurracy(Y, Y_prime):
    counter = 0
    for i, item in enumerate(Y_prime): 
          predicted_value = 0
          if item > 0.5:
               predicted_value = 1
          
          if Y[i] == predicted_value:
               counter+=1
               
    return counter/len(Y)
```

```python
x_digits, y_digits = datasets.load_digits(return_X_y=True)
     
first_class_indexes = y_digits == 3
second_class_indexes =  y_digits == 6

all_filters = first_class_indexes | second_class_indexes

#Filter dataset and scale features
x_filtered = x_digits[all_filters] 
y_filtered = y_digits[all_filters] 

binarizer = LabelBinarizer()
y_filtered = binarizer.fit_transform(y_filtered)
y_filtered = y_filtered.flatten() 

train_size = int(len(x_filtered) * 0.8)
x_train = x_filtered[0:train_size]
y_train = y_filtered[0:train_size]
x_test = x_filtered[train_size:]
y_test = y_filtered[train_size:]
```

# Singe layer implementation 

```python
import tensorflow as tf
from dense_layer import Layer, train_single_layer
```

```python
def calculate_accuracy(predicted, y):
  
  #Gets the index for the higher probability in the predicted response
  predicted = np.argmax(predicted, axis=1)
  # Find the true index on the one hot encoded ground truth
  Y = np.argmax(y, axis=1)
  
  #Creates a filter for accuracy
  indexes = predicted == Y
  
  #Returns the percentage of instances that were correctely predicted 
  return len(predicted[indexes])/len(predicted)
```

```python
def softmax(activations):
  # Calculate the exp for every items in the vector of activations
  max_act= tf.reduce_max(activations, axis=1)
  max_act = tf.reshape(max_act, (max_act.shape[0], 1))

  # Calculate the exp for every items in the vector of activations
  e = tf.math.exp(activations - max_act)
  # Get the sum of exp of activations for the entire layer
  esum = tf.reduce_sum(e, axis=1)
  # Reshape it to have the chape (m, 1 instead of (m,))
  esum = tf.reshape(esum,(e.shape[0], 1))
  
  #Divide every value of e by the total sum of esum to get the probability for each class per instance
  return e/esum
```

```python
x_digits, y_digits = datasets.load_digits(return_X_y=True)
x_digits = tf.convert_to_tensor(x_digits, dtype=tf.float32)
y_digits = tf.one_hot(y_digits, 10)

test_size = 10
x_digits, x_test = tf.split(x_digits, [int(len(x_digits)-test_size), test_size])
y_digits, y_test = tf.split(y_digits, [int(len(y_digits)-test_size), test_size])

denseLayer = Layer((10, 64), activation=softmax)


history = train_single_layer(x_digits, y_digits, denseLayer, epochs=300, learning_rate=0.001 , calculate_accuracy=calculate_accuracy)

print(f"Accuracy: {np.max(history['validation']['accuracy'])}")
```

```python
plot_loss_accuracy(history, 300)
```
